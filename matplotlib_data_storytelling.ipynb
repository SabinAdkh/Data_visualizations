### Matplotlib Object-Oriented method

>To create a graph using the OO interface, we use the plt.subplots() function, which generates an empty plot and returns a tuple of two objects:

```python
plt.subplot()
```

```python
fig, ax = plt.subplots()
print(type(fig))
print(type(ax))
```

```python
fig, ax = plt.subplots(figsize=(8, 8))
ax.barh(['A', 'B', 'C'], [2, 3, 4])
```

```python
fig, ax = plt.subplots(figsize=(8, 8))
ax.barh(['A', 'B', 'C'], [2, 3, 4])
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
```

```python
fig, ax = plt.subplots(figsize=(8, 8))
ax.barh(['A', 'B', 'C'], [2, 3, 4])

# Removing all the axes(spines)
for spine in ['left', 'right', 'top', 'bottom']:
    ax.spines[spine].set_visible(False)
```

```python
fig, ax = plt.subplots(figsize=(6, 6))
ax.barh(['A', 'B', 'C'], [2, 3, 4])

# Removing ticks from X, and y axes
ax.tick_params(bottom=False, left=False)
```

```python
fig, ax = plt.subplots()
# Decrese the thickness of the bar
ax.barh(['A', 'B', 'C'], [2, 3, 4], height=0.5)
```

```python
fig, ax = plt.subplots()
ax.barh(['A', 'B', 'C'], [2, 3, 4], height=0.5)

# Remove some xticks
ax.set_xticks([0, 1, 2, 3, 4])
```

```python
fig, ax = plt.subplots()
ax.barh(['A', 'B', 'C'], [2, 3, 4], height=0.5)

# Placing xticks at the top for easier reading
ax.xaxis.tick_top()
ax.set_xticks([0, 1, 2, 3, 4])
ax.tick_params(left = False, top=False)

for spine in ['left', 'right', 'top', 'bottom']:
    ax.spines[spine].set_visible(False)
```

```python
fig, ax = plt.subplots()
ax.barh(['A', 'B', 'C'], [2, 3, 4], height=0.5, color='#8a042a')
for spine in ['left', 'right', 'top', 'bottom']:
    ax.spines[spine].set_visible(False)
    
ax.xaxis.tick_top()
ax.set_xticks([0, 1, 2, 3, 4])
ax.tick_params(top=False, left=False)
# change the color of x_ticks to grey
ax.tick_params(axis='x', colors='grey')
```

```python
fig, ax = plt.subplots()
ax.bar(['A', 'B', 'C'],
       [2, 4, 16])
ax.text(x=0.5, y=18, s='Title example')
ax.text(x=0.5, y=17, s='Subtitle example')
plt.show()
```

```python
fig, ax = plt.subplots()
ax.bar(['A', 'B', 'C'],
       [2, 4, 16])

ax.text(x=0.5, y=18, s='Title example', size=15, weight='bold')
ax.text(x=0.5, y=17, s='Subtitle example')
```

```python
fig, ax = plt.subplots()
ax.barh(['A', 'B', 'C'], [2, 3, 4], height=0.5)

# Placing xticks at the top for easier reading
ax.xaxis.tick_top()
ax.set_xticks([0, 1, 2, 3, 4])
ax.set_xticklabels(['0.0', '1.0', '2.0', '3.0', '4.0'])
ax.tick_params(left = False, top=False)

for spine in ['left', 'right', 'top', 'bottom']:
    ax.spines[spine].set_visible(False)
```
